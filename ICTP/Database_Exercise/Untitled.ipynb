### Carbono Negro Equivalente

En esta sección se encuentran los resultados del análisis de carbono negro equivalente (CNe) presente en las muestras de PM2.5 colectadas por la Red Manual de Monitoreo Atmosférico entre 2015 y 2017, y analizadas con un transmisómetro MAGEE OT-21; así como las concentraciones horarias de CNe obtenidas de la medición continua realizada en las instalaciones de 

* Laboratorio de Análisis Ambiental de la Dirección de Monitoreo Atmosférico (lat: 19.483781, long: -99.147312) 

entre 2015 y 2017 para aerosoles con un punto de corte de 1 µm (PM1), empleando un analizador PAX (Droplet Measurement Technologies).

Todas las concentraciones se reportan en microgramos por metro cúbico (µg/m3).

Las concentraciones de CNe se calcularon empleando las constantes correspondientes recomendadas por el fabricante de cada instrumento.
