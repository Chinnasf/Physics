```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sstyle = "seaborn-poster"
plt.style.use(sstyle)
plt.rc('font',family = 'serif')

# Lowpass : Stokes Filter
# Bandpass : Anti-Stokes Filter
```

```python
def read_data(step,filename):
    df = (pd.read_csv(f"data/Step{step}/{filename}", header=None)
         .rename(columns={0:"wavelength",1:"intensity"}))
    return df
```

```python
def step_2(savefig=False):
    bp = read_data(2,"BE_bandpass.asc")
    lp = read_data(2,"BE_lowpass.asc")
    plt.plot(bp.wavelength,bp.intensity, label = "Stokes Filter", lw = 1.7);
    plt.plot(lp.wavelength,lp.intensity, label = "Anti-Stokes Filter", lw = 1.7);
    plt.xlabel("Wavelength [nm]")
    plt.ylabel("Intensity [counts]")
    plt.legend()
    plt.grid()
    if savefig:
        plt.savefig('images/step_2_fig.pdf')
    plt.show()
```

```python
def step_3(case,savefig=False):
    # Analysis done to probe 1
    if case == "bp":
        df1 = read_data(3,"P1_bandpass_25.asc")
        df2 = read_data(3,"P1_bandpass_115.asc")
        llabel = "Stokes Filter"
    else:
        df1 = read_data(3,"P1_lowpass_25.asc")
        df2 = read_data(3,"P1_lowpass_115.asc")
        llabel = "Anti-Stokes Filter"
    plt.plot(df1.wavelength,df1.intensity, label = f"{llabel} 25$\\degree$", lw = 1.7);
    plt.plot(df2.wavelength,df2.intensity, label = f"{llabel} 115$\\degree$", lw = 1.7);
    plt.xlabel("Wavelength [nm]")
    plt.ylabel("Intensity [counts]")
    plt.legend()
    plt.grid()
    if savefig:
        plt.savefig(f'images/step_3_{case}_fig.pdf')
    plt.show()
```

```python
def step_4(case,probe,savefig=False):
    print(f"\nShowing data of probe {probe}")
    if case == "bp":
        df1 = read_data(4,f"P{probe}_bandpass_25.asc")
        df2 = read_data(4,f"P{probe}_bandpass_115.asc")
        llabel = "Stokes Filter"
    else:
        df1 = read_data(4,"P1_lowpass_25.asc")
        df2 = read_data(4,"P1_lowpass_115.asc")
        llabel = "Anti-Stokes Filter"
    plt.plot(df1.wavelength,df1.intensity, label = f"{llabel} 25$\\degree$", lw = 1.7);
    plt.plot(df2.wavelength,df2.intensity, label = f"{llabel} 115$\\degree$", lw = 1.7);
    plt.xlabel("Wavelength [nm]")
    plt.ylabel("Intensity [counts]")
    plt.legend()
    plt.grid()
    if savefig:
        plt.savefig(f'images/step_4_{case}_P{probe}_fig.pdf')
    plt.show()
```
