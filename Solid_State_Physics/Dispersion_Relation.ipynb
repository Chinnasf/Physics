##  Dispersion Relation
Linear Chain with Two Atoms


```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

sstyle = "seaborn-poster"
plt.style.use(sstyle)
plt.rc('font',family = 'serif')

sp.init_printing(use_unicode=True)
```

```python
ωp = lambda q: np.sqrt(3/2 + np.sqrt( 9/4 - 2*(np.sin(x)**2)))
ωn = lambda q: np.sqrt(3/2 - np.sqrt( 9/4 - 2*(np.sin(x)**2)))
```

```python
x = np.linspace(-1,2,100)

plt.plot(x,ωp(x), label = "Optical")
plt.plot(x,ωn(x), label = "Acoustical")
plt.ylabel("ω/[f/M]$^{1/2}$")
plt.xlabel("qa/π")
plt.axvspan(-1, 1, color='green', alpha=0.15, lw=0)
plt.legend()
plt.grid()
plt.show()
```

Dispersion Relation

$$
    \omega ^ 2 = \frac{f}{M}\left[ \frac{3}{2} \pm \sqrt{\frac{9}{4} - 2\sin^2\left( \frac{qa}{2} \right)}\right]
$$

```python
a,q,f,M = sp.symbols('a,q,f,M')
# Acoustic branch
ω_α = sp.sqrt((f/M)*(3/2 - sp.sqrt((9/4) - 2*(sp.sin(q*a/2)**2)) )) 
gv  = sp.diff(ω_α, q) 
```

```python
gv
```

```python
gv.subs({q:0})
```

```python
gv.subs({q:sp.pi/a})
```

```python
gv.subs({q:-sp.pi/a})
```
