```python
import numpy as np
import matplotlib.pyplot as plt
```

### Sucesiones de Puntos: convergencia

$$
    P_n = (e^{-n/4}\cdot\cos(n), \:e^{-n/4}\cdot\sin(n))
$$

Espiral logaritmica con ecuación $r=e^{-\theta/4}$

```python
n = lambda γ:  np.arange(-γ*np.pi,γ*np.pi)
X = lambda γ: [np.exp(-i/4)*np.cos(i) for i in n(γ)]
Y = lambda γ: [np.exp(-i/4)*np.sin(i) for i in n(γ)]
```

```python
plt.plot(X(2),Y(2));
plt.scatter(X(2),Y(2), label=f"$\\gamma=${2}");

plt.plot(X(3),Y(3));
plt.scatter(X(3),Y(3), label=f"$\\gamma=${3}");

plt.grid()
plt.legend();
```

```python
for j in range(1,6):
    plt.plot(X(j),Y(j));
    plt.scatter(X(j),Y(j), label=f"$\\gamma=${j}");

plt.grid()
plt.legend();
```
